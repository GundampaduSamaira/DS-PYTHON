```python
print("------Welcome to List and its Methods in Python------")
print()
print("Initializing Lists: ")
print()
l1 = ['python',45,69.3,12,'film',12,4,65.6,1,3]
print("Initializing a list: ",l1)
l2 = list()
l3 = []
print("Initializing empty List using list() function: ",l2)
print("Initializing empty list with [] : ",l3)
print()

print("List Methods: ")
print()
lst = [10,34,62,89.4,'movie',10,32]
print("Let the List lst be --> [10,34,62,89.4,'movie',10,32]")
print()

print("append() method:")
lst.append(90.5)
print("After lst.append(90.5) the updated list is: ",lst)
print()
print("extend() method:")
lst.extend([10,100,33])
print("After lst.extend([10,100,33]) the updated list is: ",lst)
print()
print("insert() method:")
lst.insert(5,12)
print("After lst.insert(5,12) the updated list is: ",lst)
print()
print("remove() method:")
lst.remove(10)
print("After lst.remove(10) the updated list is: ",lst)
print()
print("pop() method:")
lst.pop()
print("After lst.pop() the updated list is: ",lst)
lst.pop(4)
print("After lst.pop(4) the updated list is: ",lst)
print()
print("deleting elements using del keyword:")
del lst[2:4]
print("After del lst[2:4] the updated list is: ",lst)
del lst[5]
print("After del lst[5] the updated list is: ",lst)
print("del lst deletes the entire list")
print()
print("clear() method: ")
lst.clear()
print("After lst.clear() the updated list is: ",lst)
print()
l4 = [67, 83, 99, 25, 60, 33, 12, 774, 99, 45, 12, 99]
print("Let the List l4 be --> [67, 83, 99, 25, 60, 33, 12, 774, 99, 45, 12, 99]")
print("count() method: ")
print("After l4.count(99) output is: ",l4.count(99))
print()
print("len() method: ")
print("After len(l4) output is: ",len(l4))
print()
print("min() method: ")
print("After min(l4) output is: ",min(l4))
print()
print("max() method: ")
print("After max(l4) output is: ",max(l4))
print()
print("sum() method: ")
print("After sum(l4) output is: ",sum(l4))
print()
print("sort() method: ")
print("Sort method sorts the list in place and Default sort is ascending order.")
l4.sort()
print("After l4.sort() output is: ",l4)
l4.sort(reverse = True)
print("After l4.sort(reverse = True) output is: ",l4)
print()
print("reverse() method: ")
print("reverse method reverses the original list.")
l5 = [1, 2, 'code', 'event', 3, 'f1', 2, 'hi', 32]
print("Let the list l5 be --> ",l5)
l5.reverse()
print("After l5.reverse() output is: ",l5)
print()
print("sorted() method: ")
print("creates a new sorted list of original list and default sort is ascending order.")
l6 = [90,5.6,34,31,9.0,71]
print("Let the list l6 be --> ",l6)
l7 = sorted(l6)
print("After l7 = sorted(l6) the output is: ",l7)
print("id of original list l6: ",id(l6))
print("id of new sorted list l7: ",id(l7))
print()
print("------END------")
```
